```python
## generate periods

from datetime import datetime, timedelta
start_dt = datetime.strptime('2020-10-19', '%Y-%m-%d')
periods = [start_dt + timedelta(days=7*per) for per in range(0, 52)]
```

```python
periods_df = pd.DataFrame({'Date': periods})
periods_stocks_df = pd.merge(periods_df, mf_pd, how='left', on='Date')
periods_stocks_df.head()
```

```python
import numpy as np

periods_stocks_df.set_index('Date', inplace=True)
periods_stocks_df.replace(b'', np.nan, inplace=True) # replace b-literal to np.nan
periods_stocks_df.head()
```
